# Image Convolution and Edge Detection Using Sobel OPerator

### Image Convolution
###  Convolution in image processing is a mathematical operation applied to an image to extract features such as edges, textures, and patterns.
### Process: It involves overlaying a small matrix (called a kernel or filter) on the image and performing element-wise multiplication, summing up the results to produce a single value. This is repeated across the entire image.
### Purpose: Convolution enhances certain aspects of an image by amplifying specific patterns like edges, corners, and gradients, making it crucial in feature extraction for computer vision.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('ep2.jfif', cv2.IMREAD_GRAYSCALE)

# Apply Sobel filter in the x direction
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_x = cv2.convertScaleAbs(sobel_x)  # Convert to uint8 for displaying

# Apply Sobel filter in the y direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_y = cv2.convertScaleAbs(sobel_y)  # Convert to uint8 for displaying

# Combine both directions to get edge magnitude
sobel_combined = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)

# Plot the images
plt.figure(figsize=(12, 8))
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel - X direction')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel - Y direction')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel - Combined')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('sharp.jfif', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian Blur to the original image
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 1.4)


# Combine both directions to get edge magnitude
sobel_combined = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)

# Plot the images
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(gaussian_blur, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Apply Gabor filters at different orientations (0, 45, 90, and 135 degrees)
# Define the parameters for the Gabor filter
ksize = 31  # Size of the filter
sigma = 4.0  # Standard deviation of the Gaussian function
lambd = 10.0  # Wavelength of the sinusoidal factor
gamma = 0.5  # Spatial aspect ratio
psi = 0  # Phase offset

# Create and apply Gabor filters
angles = [0, 45, 90, 135]
gabor_images = []

for theta in angles:
    theta_rad = theta * np.pi / 180  # Convert angle to radians
    gabor_kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta_rad, lambd, gamma, psi, ktype=cv2.CV_64F)
    gabor_image = cv2.filter2D(gaussian_blur, cv2.CV_8UC3, gabor_kernel)
    gabor_images.append(gabor_image)

# Plot the original blurred image and Gabor-filtered images
plt.figure(figsize=(12, 8))

plt.subplot(1, 5, 1)
plt.imshow(gaussian_blur, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')

for i, (angle, gabor_image) in enumerate(zip(angles, gabor_images), start=2):
    plt.subplot(1, 5, i)
    plt.imshow(gabor_image, cmap='gray')
    plt.title(f'Gabor Filter {angle}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Apply Laplacian edge detection on the Gaussian-blurred image
laplacian = cv2.Laplacian(gaussian_blur, cv2.CV_64F, ksize=3)
laplacian = cv2.convertScaleAbs(laplacian)  # Convert to uint8 for display

# Plot the original Gaussian-blurred image and the Laplacian edge-detected image
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 2)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Apply FFT to the original image
dft = cv2.dft(np.float32(image), flags=cv2.DFT_COMPLEX_OUTPUT)

# Shift the zero frequency component to the center of the spectrum
dft_shifted = np.fft.fftshift(dft)

# Compute the magnitude spectrum
magnitude_spectrum = cv2.magnitude(dft_shifted[:, :, 0], dft_shifted[:, :, 1])

# Normalize the magnitude spectrum to the range [0, 255]
magnitude_spectrum = np.log(1 + magnitude_spectrum)  # Log scale for better visibility
magnitude_spectrum = cv2.normalize(magnitude_spectrum, None, 0, 255, cv2.NORM_MINMAX)
magnitude_spectrum = np.uint8(magnitude_spectrum)

# Plot the original image and its magnitude spectrum
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum')
plt.axis('off')

plt.tight_layout()
plt.show()
```
